```python
wildfires_cleaned=fire[["SOURCE_REPORTING_UNIT_NAME","FIRE_NAME","FIRE_YEAR","DISCOVERY_DATE","DISCOVERY_DOY","STAT_CAUSE_DESCR","CONT_DATE","CONT_DOY","FIRE_SIZE","FIRE_SIZE_CLASS","LATITUDE","LONGITUDE","STATE","COUNTY","Shape"]]
wildfires_cleaned.head()
```

```python
fires = wildfires_cleaned.rename(columns={"SOURCE_REPORTING_UNIT_NAME":"UNIT_NAME", "STAT_CAUSE_DESCR":"CAUSE_DESCR","FIRE_SIZE_CLASS":"FIRE_CLASS"})
fires.head()
```

```python
#Create a new function:
def num_missing(x):
    return sum(x.isnull())

#Applying per column:
print("Missing values per column:")
## Check how many are missing by column, and then check which ones have any missing values
print(fires.apply(num_missing, axis=0).where(lambda x : x != 0).dropna().head(20)) 
#axis=0 defines that function is to be applied on each column
```

```python
#remove FIRE_NAME because it is not that important to name a fire and many obs are missing!
del fires ["FIRE_NAME"]
```

```python
fires.head()
```

```python
print(fires.apply(num_missing, axis=0).where(lambda x : x != 0).dropna().head(20)) 
```

```python
#Julian Calendar conversion
fires['DISC_DATE'] = pd.to_datetime(fires['DISCOVERY_DATE'] - pd.Timestamp(0).to_julian_date(), unit='D')
fires.head()
```

```python
fires['CONTAIN_DATE'] = pd.to_datetime(fires['CONT_DATE'] - pd.Timestamp(0).to_julian_date(), unit='D')
fires.head()
```

```python
del fires ["DISCOVERY_DATE"]
del fires ["CONT_DATE"]
```

```python
#Deleted the Julian calendar columns for Discovery date and Contaned date which were converted to normal datetime format in DISC_DATE and CONTAIN_DATE
fires.head()
```

```python
#Missing values
#deleting the COUNTY column because of 37.7% missing values
del fires["COUNTY"]
```

```python
fires.to_sql('wildfirecleaned', con, if_exists='replace', index=False)
pd.read_sql('select * from wildfirecleaned', con)
```
